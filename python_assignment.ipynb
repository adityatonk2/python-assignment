```python
# Q5. A year is a leap year if it is divisible by 4, except that years divisible by 100 are not leap years unless they
# are also divisible by 400. Write a program that asks user for a year and print whether it is a leap year or not?
user_input_year = int(input("Enter your year"))
if user_input_year % 4 == 0:
    if user_input_year % 100 == 0:
        if user_input_year % 400 == 0:
            print("leap year")
        else:
            print("not a leap year")
    else:
        print("leap year")
else:
    print("not a leap year")
```

```python
# Q7. Write a program to find second largest element in a list ‘Numlist’. Prompt user to enter five numbers in list.
Numlist = []
for i in range(5):
    num = int(input("Enter a number:"))
    Numlist.append(num)
Numlist.sort()
print("second largest number is ",Numlist[-2])

```
